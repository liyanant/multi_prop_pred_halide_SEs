```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
K_VRH_ML_train_df = pd.read_csv('ML/K_VRH/K_VRH_TRAIN_MAE.csv')
K_VRH_ML_train_df
```

```python
K_VRH_ML_train_df.describe()
```

```python
K_VRH_ML_train_df = K_VRH_ML_train_df[['catboost', 'LGBM', 'xgboost']]
K_VRH_ML_train_df
```

```python
K_VRH_skorch_train_df = pd.read_csv('Skorch_NN/K_VRH/K_VRH_TRAIN_MAE.csv')
K_VRH_skorch_train_df
```

```python
K_VRH_skorch_train_df.describe()
```

```python
K_VRH_skorch_train_df.columns = ['skorch_1', 'skorch_2', 'skorch_3']
K_VRH_skorch_train_df
```

```python
K_VRH_train_df = pd.concat([K_VRH_ML_train_df, K_VRH_skorch_train_df], axis = 1)
K_VRH_train_df
```

```python
K_VRH_ML_test_df = pd.read_csv('ML/K_VRH/K_VRH_HOLDOUT_MAE.csv')
K_VRH_ML_test_df
```

```python
K_VRH_ML_test_df.describe()
```

```python
K_VRH_ML_test_df = K_VRH_ML_test_df[['catboost', 'LGBM', 'xgboost']]
K_VRH_ML_test_df
```

```python
K_VRH_skorch_test_df = pd.read_csv('Skorch_NN/K_VRH/K_VRH_HOLDOUT_MAE.csv')
K_VRH_skorch_test_df
```

```python
K_VRH_skorch_test_df.describe()
```

```python
K_VRH_skorch_test_df.columns = ['skorch_1', 'skorch_2', 'skorch_3']
K_VRH_skorch_test_df
```

```python
K_VRH_test_df = pd.concat([K_VRH_ML_test_df, K_VRH_skorch_test_df], axis = 1)
K_VRH_test_df
```

```python
labels = K_VRH_test_df.columns

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(30, 10))

plt.rcParams.update({'font.size': 22, 'font.weight':'bold'})
bplot1 = ax1.boxplot(K_VRH_train_df,
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=labels)  # will be used to label x-ticks
ax1.set_title('Training Rectangular box plot', fontsize = 24, fontweight='bold')


bplot2 = ax2.boxplot(K_VRH_test_df,
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=labels)  # will be used to label x-ticks
ax2.set_title('Holdout Rectangular box plot', fontsize = 24, fontweight='bold')


# fill with colors
colors = ['red', 'blue', 'orange', 'violet', 'yellow', 'green', 'pink', 'brown', 'lightblue', 'magenta', 'lightgreen', 'cyan', 'beige']
for bplot in (bplot1, bplot2):
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)


# adding horizontal grid lines
for ax in [ax1, ax2]:
    ax.yaxis.grid(False)
    ax.xaxis.grid(False)
    ax.set_ylabel('MAE (GPa)', fontsize = 24, fontweight='bold')

plt.show()
```

```python
from matplotlib.ticker import FormatStrFormatter
```

```python
labels = K_VRH_test_df.columns

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(30, 10))

plt.rcParams.update({'font.size': 22, 'font.weight':'bold'})
bplot1 = ax1.boxplot(K_VRH_train_df,
                     vert=True,
                     patch_artist=True,
                     labels=labels)
ax1.set_title('Training Rectangular box plot', fontsize = 24, fontweight='bold')
ax1.yaxis.set_major_formatter(FormatStrFormatter('%.1f'))

bplot2 = ax2.boxplot(K_VRH_test_df,
                     vert=True,
                     patch_artist=True,
                     labels=labels)
ax2.set_title('Holdout Rectangular box plot', fontsize = 24, fontweight='bold')
ax2.yaxis.set_major_formatter(FormatStrFormatter('%.1f'))

colors = ['red', 'blue', 'orange', 'violet', 'yellow', 'green', 'pink', 'brown', 'lightblue', 'magenta', 'lightgreen', 'cyan', 'beige']
for bplot in (bplot1, bplot2):
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)

for ax in [ax1, ax2]:
    ax.yaxis.grid(False)
    ax.xaxis.grid(False)
    ax.set_ylabel('MAE (GPa)', fontsize = 24, fontweight='bold')

plt.show()
```
