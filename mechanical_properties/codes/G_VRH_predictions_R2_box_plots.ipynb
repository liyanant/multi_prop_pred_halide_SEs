```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
G_VRH_trad_train_df = pd.read_csv('Trad_ML/G_VRH/G_VRH_TRAIN_R2.csv')
G_VRH_trad_train_df
```

```python
G_VRH_trad_train_df.describe()
```

```python
G_VRH_trad_train_df = G_VRH_trad_train_df[['catboost', 'LGBM', 'xgboost']]
G_VRH_trad_train_df
```

```python
G_VRH_skorch_train_df = pd.read_csv('Skorch_NN/G_VRH/G_VRH_TRAIN_R2.csv')
G_VRH_skorch_train_df
```

```python
G_VRH_skorch_train_df.describe()
```

```python
G_VRH_skorch_train_df.columns = ['skorch_1', 'skorch_2', 'skorch_3']
G_VRH_skorch_train_df
```

```python
G_VRH_train_df = pd.concat([G_VRH_trad_train_df, G_VRH_skorch_train_df], axis = 1)
G_VRH_train_df
```

```python
G_VRH_trad_test_df = pd.read_csv('Trad_ML/G_VRH/G_VRH_HOLDOUT_R2.csv')
G_VRH_trad_test_df
```

```python
G_VRH_trad_test_df.describe()
```

```python
G_VRH_trad_test_df = G_VRH_trad_test_df[['catboost', 'LGBM', 'xgboost']]
G_VRH_trad_test_df
```

```python
G_VRH_skorch_test_df = pd.read_csv('Skorch_NN/G_VRH/G_VRH_HOLDOUT_R2.csv')
G_VRH_skorch_test_df
```

```python
G_VRH_skorch_test_df.describe()
```

```python
G_VRH_skorch_test_df.columns = ['skorch_1', 'skorch_2', 'skorch_3']
G_VRH_skorch_test_df
```

```python
G_VRH_test_df = pd.concat([G_VRH_trad_test_df, G_VRH_skorch_test_df], axis = 1)
G_VRH_test_df
```

```python
labels = G_VRH_test_df.columns

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(30, 10))

plt.rcParams.update({'font.size': 22, 'font.weight':'bold'})
bplot1 = ax1.boxplot(G_VRH_train_df,
                     vert=True,
                     patch_artist=True,
                     labels=labels)
ax1.set_title('Training Rectangular box plot', fontsize = 24, fontweight='bold')


bplot2 = ax2.boxplot(G_VRH_test_df,
                     vert=True,
                     patch_artist=True,
                     labels=labels)
ax2.set_title('Holdout Rectangular box plot', fontsize = 24, fontweight='bold')

colors = ['red', 'blue', 'orange', 'violet', 'yellow', 'green', 'pink', 'brown', 'lightblue', 'magenta', 'lightgreen', 'cyan', 'beige']
for bplot in (bplot1, bplot2):
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)

for ax in [ax1, ax2]:
    ax.yaxis.grid(False)
    ax.xaxis.grid(False)
    ax.set_ylabel('$\mathregular{R^2}$', fontsize = 24, fontweight='bold')

plt.show()
```
