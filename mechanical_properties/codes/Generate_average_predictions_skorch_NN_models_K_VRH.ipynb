```python
df1 = pd.read_csv('NEW_K_VRH_predictions_for_SCREENED_Li-based_Na_based_K-based_Mg-based_Ba-based_Ca-based_and_Al-based_compounds_from_MPJ_with_no_mech_prop_info.csv')
df1
```

```python
skorch_model_1 = df1[df1.columns[0:100]]
skorch_model_2 = df1[df1.columns[100:200]]
skorch_model_3 = df1[df1.columns[200:300]]
```

```python
model_list_names = ['skorch_model_1', 'skorch_model_2', 'skorch_model_3']
```

```python
model_list = [skorch_model_1, skorch_model_2, skorch_model_3]
```

```python
for i in range(len(model_list)):
    print(model_list[i].columns)
    print()
    print('-------------------')
    print()
```

```python
for i in range(len(model_list)):
    model_list[i][model_list_names[i] + '_Average_prediction'] = model_list[i].mean(axis=1)
    print(model_list[i][[model_list_names[i] + '_Average_prediction']])
    print()
    print('-------------------')
    print()
```

```python
for i in range(len(model_list)):
  model_list[i].to_csv(model_list_names[i] + '.csv', index=False)
```
