```python
df = pd.read_csv('No_mech_prop_data_from_paper_of_SCREENED_Li-based_Na_based_K-based_Mg-based_Ba-based_Ca-based_and_Al-based_compounds_from_MPJ_16_July_2025.csv')
df
```

```python
df1 = pd.read_csv('NEW_Average_skorch_NN_model_predictions_for_K_VRH.csv')
df1
```

```python
df2 = pd.read_csv('NEW_Average_skorch_NN_model_predictions_for_G_VRH.csv')
df2
```

```python
new_df = pd.concat([df, df1[['skorch_model_1_Average_prediction', 'skorch_model_2_Average_prediction', 'skorch_model_3_Average_prediction']]], axis = 1)
new_df
```

```python
new_df[['skorch_model_1_K_VRH_Average_prediction', 'skorch_model_2_K_VRH_Average_prediction', 'skorch_model_3_K_VRH_Average_prediction']] = new_df[['skorch_model_1_Average_prediction', 'skorch_model_2_Average_prediction', 'skorch_model_3_Average_prediction']]
new_df
```

```python
new_df = new_df.drop(['skorch_model_1_Average_prediction', 'skorch_model_2_Average_prediction', 'skorch_model_3_Average_prediction'], axis = 1)
new_df
```

```python
new_df = pd.concat([new_df, df2[['skorch_model_1_Average_prediction', 'skorch_model_2_Average_prediction', 'skorch_model_3_Average_prediction']]], axis = 1)
new_df
```

```python
new_df[['skorch_model_1_G_VRH_Average_prediction', 'skorch_model_2_G_VRH_Average_prediction', 'skorch_model_3_G_VRH_Average_prediction']] = new_df[['skorch_model_1_Average_prediction', 'skorch_model_2_Average_prediction', 'skorch_model_3_Average_prediction']]
new_df
```

```python
new_df = new_df.drop(['skorch_model_1_Average_prediction', 'skorch_model_2_Average_prediction', 'skorch_model_3_Average_prediction'], axis = 1)
new_df
```

```python
new_df = new_df[new_df['skorch_model_1_G_VRH_Average_prediction'] > 8.5]
new_df = new_df[new_df['skorch_model_2_G_VRH_Average_prediction'] > 8.5]
new_df = new_df[new_df['skorch_model_3_G_VRH_Average_prediction'] > 8.5]
```

```python
new_df
```

```python
new_df['skorch_model_1_Pugh_ratio'] = new_df['skorch_model_1_K_VRH_Average_prediction'] / new_df['skorch_model_1_G_VRH_Average_prediction']
new_df
```

```python
new_df['skorch_model_2_Pugh_ratio'] = new_df['skorch_model_2_K_VRH_Average_prediction'] / new_df['skorch_model_2_G_VRH_Average_prediction']
new_df
```

```python
new_df['skorch_model_3_Pugh_ratio'] = new_df['skorch_model_3_K_VRH_Average_prediction'] / new_df['skorch_model_3_G_VRH_Average_prediction']
new_df
```

```python
new_df = new_df[new_df['skorch_model_1_Pugh_ratio'] >= 1.75]
new_df = new_df[new_df['skorch_model_2_Pugh_ratio'] >= 1.75]
new_df = new_df[new_df['skorch_model_3_Pugh_ratio'] >= 1.75]
new_df
```
