```python
df = pd.read_csv('mech_prop_prediction_dataset.csv')
df
```

```python
df1 = pd.read_csv('NEW_Average_ML_model_predictions_for_K_VRH.csv')
df1
```

```python
df2 = pd.read_csv('NEW_Average_ML_model_predictions_for_G_VRH.csv')
df2
```

```python
new_df = pd.concat([df, df1[['catboost_Average_prediction', 'LGBM_Average_prediction', 'xgboost_Average_prediction']]], axis = 1)
new_df
```

```python
new_df[['catboost_K_VRH_Average_prediction', 'LGBM_K_VRH_Average_prediction', 'xgboost_K_VRH_Average_prediction']] = new_df[['catboost_Average_prediction', 'LGBM_Average_prediction', 'xgboost_Average_prediction']]
new_df
```

```python
new_df = new_df.drop(['catboost_Average_prediction', 'LGBM_Average_prediction', 'xgboost_Average_prediction'], axis = 1)
new_df
```

```python
new_df = pd.concat([new_df, df2[['catboost_Average_prediction', 'LGBM_Average_prediction', 'xgboost_Average_prediction']]], axis = 1)
new_df
```

```python
new_df[['catboost_G_VRH_Average_prediction', 'LGBM_G_VRH_Average_prediction', 'xgboost_G_VRH_Average_prediction']] = new_df[['catboost_Average_prediction', 'LGBM_Average_prediction', 'xgboost_Average_prediction']]
new_df
```

```python
new_df = new_df.drop(['catboost_Average_prediction', 'LGBM_Average_prediction', 'xgboost_Average_prediction'], axis = 1)
new_df
```

```python
new_df = new_df[new_df['catboost_G_VRH_Average_prediction'] > 8.5]
new_df = new_df[new_df['LGBM_G_VRH_Average_prediction'] > 8.5]
new_df = new_df[new_df['xgboost_G_VRH_Average_prediction'] > 8.5]
```

```python
new_df
```

```python
new_df['catboost_Pugh_ratio'] = new_df['catboost_K_VRH_Average_prediction'] / new_df['catboost_G_VRH_Average_prediction']
new_df
```

```python
new_df['LGBM_Pugh_ratio'] = new_df['LGBM_K_VRH_Average_prediction'] / new_df['LGBM_G_VRH_Average_prediction']
new_df
```

```python
new_df['xgboost_Pugh_ratio'] = new_df['xgboost_K_VRH_Average_prediction'] / new_df['xgboost_G_VRH_Average_prediction']
new_df
```

```python
new_df = new_df[new_df['catboost_Pugh_ratio'] >= 1.75]
new_df = new_df[new_df['LGBM_Pugh_ratio'] >= 1.75]
new_df = new_df[new_df['xgboost_Pugh_ratio'] >= 1.75]
new_df
```
