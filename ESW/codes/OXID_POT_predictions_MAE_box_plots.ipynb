```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
OXID_POT_ML_train_df = pd.read_csv('ML/Oxidation_potential/OXID_POT_TRAIN_MAE.csv')
OXID_POT_ML_train_df
```

```python
OXID_POT_ML_train_df.describe()
```

```python
OXID_POT_ML_train_df = OXID_POT_ML_train_df[['catboost', 'LGBM', 'xgboost']]
OXID_POT_ML_train_df
```

```python
OXID_POT_skorch_train_df = pd.read_csv('Skorch_NN/Oxidation_potential/OXID_POT_TRAIN_MAE.csv')
OXID_POT_skorch_train_df
```

```python
OXID_POT_skorch_train_df.describe()
```

```python
OXID_POT_skorch_train_df.columns = ['skorch_1', 'skorch_2', 'skorch_3']
OXID_POT_skorch_train_df
```

```python
OXID_POT_train_df = pd.concat([OXID_POT_ML_train_df, OXID_POT_skorch_train_df], axis = 1)
OXID_POT_train_df
```

```python
OXID_POT_ML_test_df = pd.read_csv('ML/Oxidation_potential/OXID_POT_HOLDOUT_MAE.csv')
OXID_POT_ML_test_df
```

```python
OXID_POT_ML_test_df.describe()
```

```python
OXID_POT_ML_test_df = OXID_POT_ML_test_df[['catboost', 'LGBM', 'xgboost']]
OXID_POT_ML_test_df
```

```python
OXID_POT_skorch_test_df = pd.read_csv('Skorch_NN/Oxidation_potential/OXID_POT_HOLDOUT_MAE.csv')
OXID_POT_skorch_test_df
```

```python
OXID_POT_skorch_test_df.describe()
```

```python
OXID_POT_skorch_test_df.columns = ['skorch_1', 'skorch_2', 'skorch_3']
OXID_POT_skorch_test_df
```

```python
OXID_POT_test_df = pd.concat([OXID_POT_ML_test_df, OXID_POT_skorch_test_df], axis = 1)
OXID_POT_test_df
```

```python
from matplotlib.ticker import FormatStrFormatter
```

```python
labels = OXID_POT_test_df.columns

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(30, 10))

plt.rcParams.update({'font.size': 22, 'font.weight':'bold'})
bplot1 = ax1.boxplot(OXID_POT_train_df,
                     vert=True,
                     patch_artist=True,
                     labels=labels)
ax1.set_title('Training Rectangular box plot', fontsize = 24, fontweight='bold')


bplot2 = ax2.boxplot(OXID_POT_test_df,
                     vert=True,
                     patch_artist=True,
                     labels=labels)
ax2.set_title('Holdout Rectangular box plot', fontsize = 24, fontweight='bold')


# fill with colors
colors = ['red', 'blue', 'orange', 'violet', 'yellow', 'green', 'pink', 'brown', 'lightblue', 'magenta', 'lightgreen', 'cyan', 'beige']
for bplot in (bplot1, bplot2):
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)

for ax in [ax1, ax2]:
    ax.yaxis.grid(False)
    ax.xaxis.grid(False)
    ax.set_ylabel('MAE (V)', fontsize = 24, fontweight='bold')

plt.show()
```
