```python
df1 = pd.read_csv('Shortlisted_Li-containing_materials_via_average_predictions_of_all_skorch_nn_models.csv')
df1
```

```python
df2 = df1[df1['F'] != 0]
df2
```

```python
df3 = df1[df1['Cl'] != 0]
df3
```

```python
df4 = df1[df1['Br'] != 0]
df4
```

```python
df6 = df1[df1['At'] != 0]
df6
```
