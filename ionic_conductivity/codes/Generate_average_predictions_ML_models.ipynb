```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('NEW-Ionic_conductivity_predictions_for_filtered_MPJ_database.csv')
df
```

```python
# ['et', 'catboost', 'RF', 'LGBM', 'ada']
et = df[df.columns[0:100]]
catboost = df[df.columns[100:200]]
RF = df[df.columns[200:300]]
LGBM = df[df.columns[300:400]]
ada = df[df.columns[400:500]]
```

```python
model_list = [et, catboost, RF, LGBM, ada]
```

```python
model_list_names = ['et', 'catboost', 'RF', 'LGBM', 'ada']
```

```python
for i in range(len(model_list)):
    print(model_list[i].columns)
    print()
    print('-------------------')
    print()
```

```python
for i in range(len(model_list)):
    model_list[i][model_list_names[i] + '_Average_prediction'] = model_list[i].mean(axis=1)
    print(model_list[i][[model_list_names[i] + '_Average_prediction']])
    print()
    print('-------------------')
    print()
```

```python
for i in range(len(model_list)):
  model_list[i].to_csv(model_list_names[i] + '.csv', index=False)
```
