```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ML_train_acc_df = pd.read_csv('ML/TRAIN_ACCURACIES.csv')
ML_train_acc_df
```

```python
ML_train_acc_df.describe()
```

```python
ML_train_acc_df = ML_train_acc_df[['et', 'catboost', 'RF', 'ada', 'LGBM']]
ML_train_acc_df
```

```python
skorch_train_acc_df = pd.read_csv('Skorch_NN/TRAIN_ACCURACIES.csv')
skorch_train_acc_df
```

```python
skorch_train_acc_df.describe()
```

```python
skorch_train_acc_df.columns = ['skorch_1', 'skorch_2', 'skorch_3']
skorch_train_acc_df
```

```python
train_acc_df = pd.concat([ML_train_acc_df, skorch_train_acc_df], axis = 1)
train_acc_df
```

```python
ML_test_acc_df = pd.read_csv('ML/HOLDOUT_ACCURACIES.csv')
ML_test_acc_df
```

```python
ML_test_acc_df.describe()
```

```python
ML_test_acc_df = ML_test_acc_df[['et', 'catboost', 'RF', 'ada', 'LGBM']]
ML_test_acc_df
```

```python
skorch_test_acc_df = pd.read_csv('Skorch_NN/HOLDOUT_ACCURACIES.csv')
skorch_test_acc_df
```

```python
skorch_test_acc_df.describe()
```

```python
skorch_test_acc_df.columns = ['skorch_1', 'skorch_2', 'skorch_3']
skorch_test_acc_df
```

```python
test_acc_df = pd.concat([ML_test_acc_df, skorch_test_acc_df], axis = 1)
test_acc_df
```

```python
labels = test_acc_df.columns

plt.rcParams.update({'font.size': 12, 'font.weight':'bold'})
plt.figure(figsize = (10,6))
bplot1 = plt.boxplot(train_acc_df,
                     vert=True,
                     patch_artist=True,
                     labels=labels)
plt.title('Training Rectangular box plot', fontsize = 18, fontweight='bold')


# fill with colours
colors = ['red', 'blue', 'orange', 'violet', 'yellow', 'green', 'pink', 'brown', 'lightblue', 'magenta', 'lightgreen', 'cyan', 'beige']


for patch, color in zip(bplot1['boxes'], colors):
    patch.set_facecolor(color)

plt.ylabel('Accuracy', fontsize = 18, fontweight='bold')

plt.show()
```

```python
labels = test_acc_df.columns

plt.rcParams.update({'font.size': 12, 'font.weight':'bold'})
plt.figure(figsize = (10,6))
bplot1 = plt.boxplot(test_acc_df,
                     vert=True,
                     patch_artist=True,
                     labels=labels)
plt.title('Holdout Rectangular box plot', fontsize = 18, fontweight='bold')


# fill with colors
colors = ['red', 'blue', 'orange', 'violet', 'yellow', 'green', 'pink', 'brown', 'lightblue', 'magenta', 'lightgreen', 'cyan', 'beige']


for patch, color in zip(bplot1['boxes'], colors):
    patch.set_facecolor(color)

plt.ylabel('Accuracy', fontsize = 18, fontweight='bold')

plt.show()
```
